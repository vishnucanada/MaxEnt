```python
import sympy as smp
```

```python
lambda_0, lambda_1, x, u = smp.symbols('lambda_0 lambda_1 x u', real=True)
a, b = smp.symbols('a b', real=True)
```

$$
  J(p) = \int_{a}^{b} p(x) \ln p(x) \, dx - \lambda_0\ (\int_{a}^{b} p(x)dx-1)
$$

$$
    \frac{dJ}{dp(x)}= 1+\ln p(x)dx-\lambda_0=0
$$

$$
    p(x)= e^{1-\lambda_0}
$$

```python
lagrange = smp.exp(1 - lambda_1)
lagrange
```

We must satisfy the constraint
$$
    \int_{a}^{b} p(x)dx = 1
$$


$$
    \int_{a}^{b} e^{1-\lambda_0}dx = 1
$$

$$
    e^{1-\lambda_0}  \int_{a}^{b}dx = 1
$$

$$
    e^{1-\lambda_0}(b-a)= 1 
$$

$$
    \lambda_0 = 1 - \ln \frac{1}{b-a}
$$

```python
integral = smp.integrate(lagrange, (x, a, b))
solution = smp.solve(integral - 1, lambda_1)
solution
```

Recall our original function was
$$
    p(x) = e^{1-\lambda_0}
$$


So plugging in $\lambda_0$
we get 
$$
    p(x) = e^{1-(1- \ln \frac{1}{b-a})}
    $$


$$
    p(x) = \frac{1}{b-a}
$$
Which is the definition of the probability density function of a unifrom diistribution
